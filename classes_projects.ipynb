```python
class Usuario:

    def __init__(self, id, path_to_user_file):
        self.id=id
        self.path_to_user_file=path_to_user_file
        self.meta={}

        with open(path_to_user_file, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                id, age, gender, occupation, code = line.split(sep='|')

                if self.id == id:
                    self.meta['age'] = age
                    self.meta['gender'] = gender
                    self.meta['occupation'] = occupation
                    self.meta['code'] = code.replace('\n','')

                    break

    def valorMedio(self, path_to_dataset):
        rating_list = []

        with open(path_to_dataset, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                user, item, rating, time_stamp = line.split(sep='\t')
  
                if user == self.id:
                    rating_list.append(float(rating))
                else:
                    pass

        self.mean = mean(rating_list)

    

class Filme:

    def __init__(self, id, path_to_item_file):
        self.id=id
        self.path_to_item_file=path_to_item_file
        self.meta={}

        with open(path_to_item_file, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                
                id, title, date, video_date, url = line.split(sep='|')[0:5]

                if self.id == id:
                    self.meta['title'] = title
                    self.meta['date'] = date
                    self.meta['video_date'] = video_date
                    self.meta['url'] = url

                    break
    

    def valorMedio(self, path_to_dataset):
        rating_list = []

        with open(path_to_dataset, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                user, item, rating, time_stamp = line.split(sep='\t')
  
                if item == self.id:
                    rating_list.append(float(rating))
                else:
                    pass

        self.mean = mean(rating_list)

    def __str__(self):
        return self.meta['title']


class Genero:
    def __init__(self, id, path_to_genre_file):
        self.id=id
        self.path_to_genre_file=path_to_genre_file
        self.meta={}

        with open(path_to_genre_file, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                genre, id = line.split(sep='|')
                id = id.replace('\n','')

                if self.id == id:
                    self.meta['genre'] = genre



class Profissao:
    def __init__(self, path_to_occuaption_file):
        self.occupation=[]

        with open(path_to_occuaption_file, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                self.occupation.append(line.replace('\n',''))

class Avaliacao:
    def __init__(self, user, path_to_dataset):
        self.user=user
        self.path_to_dataset=path_to_dataset
        self.ratings=[]

        with open(path_to_dataset, 'r', encoding = "ISO-8859-1") as f:
            for line in f.readlines():
                user, item, rating, time_stamp = line.split(sep='\t')

                if user == self.user.id:
                    self.ratings.append((Filme(item, 'ml-100k/u.item'), rating))
    

```

```python

class SistemaDeRecomendacao:
    def __init__(self, users, filmes,occupations, genres):
        self.users=users
        self.filmes=filmes
        self.occupations=occupations
        self.genres=genres
        self.data=[]

    def addUser(self, user):
        self.users.append(user)
    
    def addFilme(self, filme):
        self.filmes.append(filme)
    
    def addAvaliacao(self, avaliacao):
        self.data.append(avaliacao)
```

```python
gen = Genero('3', 'ml-100k/u.genre')
profissao = Profissao('ml-100k/u.occupation')
```

```python
user = Usuario('1', 'ml-100k/u.user')
avaliacao = Avaliacao(user, 'ml-100k/u.data')

sistema = SistemaDeRecomendacao([], [], profissao, gen)
sistema.addUser(user)
sistema.addFilme(Filme('100', 'ml-100k/u.item'))
sistema.addAvaliacao((user, filme, '2', '00001'))
```
